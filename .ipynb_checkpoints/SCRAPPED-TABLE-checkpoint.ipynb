```python
import PyPDF2
from tabula import read_pdf
import json

def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfFileReader(file)
        for page_num in range(pdf_reader.numPages):
            text += pdf_reader.getPage(page_num).extractText()
    return text

# Provide the path to your PDF file
pdf_path = 'C:/Users/HP/Downloads/TCS_internship_project/Food Calories List.pdf'

# Extract text from the PDF
body_text = extract_text_from_pdf(pdf_path)

# Read the PDF file and extract tabular data from all pages
data_frames = read_pdf(pdf_path, pages='all', output_format='json')

# Initialize an empty list to store JSON representations of tabular data
json_data_list = []

# Iterate over each DataFrame and convert it to JSON
for i, data_frame in enumerate(data_frames):
    table_data = data_frame['data']
    json_data_list.append({"table": table_data})

# Create the JSON structure with body text and tabular data
final_json_data = {
    "body": body_text,
    "table": json_data_list
}

# Convert the JSON data to a string
json_string = json.dumps(final_json_data, indent=4)

# Print the JSON string
print(json_string)
```
