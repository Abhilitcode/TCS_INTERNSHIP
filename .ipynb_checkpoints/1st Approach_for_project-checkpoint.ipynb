```python
from tabula import read_pdf
```

```python
from tabulate import tabulate
import pandas as pd
```

```python
df  = read_pdf("Food Calories List.pdf",pages=1)
```

```python
print(df)
```

```python
data_frame = pd.concat(df)
data_frame
```

```python
data_frame = data_frame.dropna(axis='columns')
data_frame
```

```python
excel_dframe = pd.read_excel("Food Calories List_removed.xlsx")
excel_dframe
```

```python
for i, (pdf_row, excel_row) in enumerate(zip(data_frame.itertuples(index=False), excel_dframe.itertuples(index=False)), start=1):
    print(f"Iteration {i}:")
    print("PDF Row:", pdf_row)
    print("Excel Row:", excel_row)
    print()
    for j, (pdf_value, excel_value) in enumerate(zip(pdf_row, excel_row), start=1):
        print(f"    Row {i}, Column {j}")
        print(f"    PDF Value: {pdf_value}")
        print(f"    Excel Value: {excel_value}")
        print()
```

```python
# import pandas as pd

# # Sample data
# pdf_df = pd.DataFrame({'A': [1, 4], 'B': [2, 5], 'C': [3, 6]})
# excel_df = pd.DataFrame({'A': [1, 4], 'B': [2, 9], 'C': [3, 6]})

# Initialize an empty list to store comparison results
comparison_results = []

# Iterate over each value in both dataframes
for i, (pdf_row, excel_row) in enumerate(zip(data_frame.itertuples(index=False), excel_dframe.itertuples(index=False)), start=1):
    for j, (pdf_value, excel_value) in enumerate(zip(pdf_row, excel_row), start=1):
        if pdf_value == excel_value:
            comparison_results.append((f"Row {i}, Column {j}", "match"))
        else:
            comparison_results.append((f"Row {i}, Column {j}", "wrong"))

# Create a new dataframe from the comparison results
comparison_df = pd.DataFrame(comparison_results, columns=["Key", "Result"])
pd.set_option('display.max_rows', None)
(comparison_df)
```
