```python
import PyPDF2
from tabula import read_pdf
import json
```

```python
def extract_pdf(pdf_path):
    text = ""
    with open(pdf_path,'rb') as file:
        pdf_reader = PyPDF2.PdfFileReader(file)
        for page_num in range(pdf_reader.numPages):
            text += pdf_reader.getPage(page_num).extractText()
        return text

pdf_path = 'C:/Users/HP/Downloads/TCS_internship_project/Food Calories List.pdf'
body_text = extract_pdf(pdf_path)
print(body_text)
```

```python
data_frames = read_pdf(pdf_path, pages='all', output_format='json')
table_data_list = []
for i, data_frame in enumerate(data_frames):
    table_data = data_frame['data']
    table_data_list.append({"table":table_data})

print(table_data_list)
```

```python
json_final_structure = {
    "body": body_text,
    "table": table_data_list
}

json_string_format = json.dumps(json_final_structure,indent=4)
print(json_string_format)
```

```python
import json
with open('data1.json', 'w') as f:
    json.dump(json_string_format, f)
```

```python
import json

# Open the JSON file
with open('data1.json') as json_file:
    data = json.load(json_file)

# Load the JSON data
data = json.loads(data)

# Loop through each page
for page_data in data['table'][0]['table']:
    my_data = page_data['table']

    # Get headings
    heading_row = my_data[0]
    headings = [cell['text'] for cell in heading_row if cell['text'] != '']

    # Extract rows
    result = []
    for row_i in range(1, len(my_data)):
        result_row = {}
        for i in range(len(my_data[row_i])):
            val = my_data[row_i][i]['text']
            if headings[i] != '':
                result_row[headings[i]] = val
        result.append(result_row)

    # Display extracted data for this page
    print("Data for page:", page_data['page_number'])
    for row in result:
        print(row)
```
